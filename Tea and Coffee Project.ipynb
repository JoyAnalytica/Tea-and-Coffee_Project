```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tea=pd.read_csv('tcoff.csv')
```

```python
tea.head(3)
```

#  What is the average taste_score?




```python
test=tea['taste_score'].mean()
print('Average Test Score :', test.round(2))
```

#  What is the average sleep duration for those with a 'High' caffeine preference?



```python
test=tea[tea['caffeine_preference']=='High']	
a=test['sleep_hours'].mean()
print('Average sleep hours :',a.round(2))
```

# What is the total monthly spend across all customers?


```python
a=tea['monthly_spend'].sum()
print('Monthly total Spend :',a)
```


#  Calculate the average stress level based on work type.




```python
group=tea.groupby('work_type')['stress_level'].mean().sort_values(ascending=False).round(2)
print('average stress',group)
print("Chart explanation example :"
"This chart shows the average stress level of tea and coffee drinkers across different work types."
    "Remote workers show the highest stress (5.03) while students show the lowest stress (4.96). "
"This suggests that work environment may influence stress more than beverage choice"
)

plt.figure(figsize=(11,7))
c=['g','y','r','m']
ax=group.plot(kind='bar',color=c,alpha=0.7)
plt.xlabel('Work Type',color='black',fontweight='bold',fontsize=14)
plt.ylabel('Stress Level',color='black',fontweight='bold',fontsize=14)

plt.title('Work type wise average stress level',color='black',fontweight='bold')
for container in ax.containers:
    ax.bar_label(container,padding=3,color='black',fontweight='bold')

plt.xticks(fontweight='bold',color='blue')
plt.yticks(fontweight='bold',color='blue')

plt.show()
```

#  Calculate the average BMI based on exercise frequency




```python
group=tea.groupby('exercise_frequency')['bmi'].mean().round(2).sort_values(ascending=False)
print('Exercise wise bmi',group)
print("Analysis Insight:"
"The analysis aimed to determine if higher exercise frequency correlates with a lower BMI. Surprisingly,"
"the data shows that individuals in the 'High' exercise category have a slightly higher average BMI (26.49) "
"compared to those in the 'Low' category (26.46). Given that the differences are marginal,"
"we can conclude that for this particular dataset, exercise frequency does not have a significant impact on reducing BMI."
)
plt.figure(figsize=(10,6))
c=['g','y','r','m']
ax=group.plot(kind='barh',color=c,alpha=0.7)
plt.xlabel('BMI',color='black',fontweight='bold',fontsize=14)
plt.ylabel('Frequency',color='black',fontweight='bold',fontsize=14)

plt.title('Exercis Frequency Wise Average BMI',color='black',fontweight='bold')
for container in ax.containers:
    ax.bar_label(container,padding=1,color='black',fontweight='bold',fontsize=8)

plt.xticks(fontweight='bold',color='blue')
plt.yticks(fontweight='bold',color='blue')
plt.show()
```


# Analyze total spending for each profession to identify which work type spends the most money.
